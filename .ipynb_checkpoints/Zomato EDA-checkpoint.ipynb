## Numerical Variables

```python
final_df.columns

#check aggrating: 
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']) #
```

```python
#EDA on Zomato app ratings

final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()

#white-not rated. 0.1 to 2.4 poor red, 3.5 to 3.4 avg 3.5 to 3.9- good yellow, 
#4.0-4.4 Very good. 4.5 to 5- excellent
```

```python
#convert into dataframe use reser_index()

final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()

#as we have don groupby wrt 0 ratings I have 2148 records, 1.8 1 record. etc
```

```python

#renaming 0 column into Rating count
ratings= final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().
                                    reset_index().rename(columns= {0:'Rating Count'})
ratings
```

### Observations

1. When ratings are between 4.5 to 4.9-----> excellent. 
2. When ratings are btw 4.0 to 4.4-----> very good. 
3. When ratings are btw 3.5 to 3.9 -----> Good 
4. When ratings are btw 3.0 to 2.5 -----> Average
5. When ratings are btw 2.4 to 0.1 -----> Poor
6. 0 stands for "not rated"

Most one is 0, that is people haven't rated app yet 

```python
ratings.head()
```

```python
#relationship btw Aggregate and Rating Count
sns.barplot(x= "Aggregate rating", y= "Rating Count", data= ratings)
```

```python
# to make it look bigger 

matplotlib.rcParams["figure.figsize"]= (12,6)
sns.barplot(x= "Aggregate rating", y= "Rating Count", data= ratings)

#here I can see that 0 is more 2K+

#here we have Gaussian distribution, we will be happy: better for our model 
```

```python
#executing heatmap 

sns.heatmap(df.isnull(), yticklabels= False, cbar= False, cmap='viridis') 
```

```python
#how to use same color wrt ratings in my bar graph 

sns.barplot(x= "Aggregate rating", y= "Rating Count", hue= "Rating color", data= ratings)
```

```python
#but here colors are not taken as it is.. we need to map these
sns.set_style('dark')
sns.barplot(x= "Aggregate rating", y= "Rating Count", hue= "Rating color", data= ratings, palette= ['white','red','orange','yellow','green','green'])

#use set_style to change background, not exactly to black 
```

### Observations

1. Not rated count is very high 
2. Maximum number of ratings are between 2.5 and 3.4
 

```python
#Missing ratings.
# if there are any missing ratings, then we can use 2.5 and 3.4 as they are maximum 
```

```python
# Count plot : plotting wrt categorical variables

sns.countplot(x= "Rating color", data= ratings, palette= ['white','red','orange','yellow','green','green'])

#once I have executed sns.set_style('dark'), then my all sns graphs will be in dark mode only 

# ***** This below is not the count: it is the frequency *****
# there is only 1 time white appeared in the list. so it is less. 
# orange has been appreared 10 times so it is highest 
```
